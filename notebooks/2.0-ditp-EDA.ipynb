```python
# Efficiently check for dtypes, non-null values, unique values, NaN values.

def summarize_df(df):
    summary = pd.DataFrame({
        'Data Type': df.dtypes,
        'Non-Null Count': df.notnull().sum(),
        'Unique Values': df.nunique(),
        'NaN Count': df.isnull().sum()
    })
    return summary


summary = summarize_df(df)
print(summary)
```

```python
df.describe().T
```

```python
# Day, month, year, hour, and weekday have already been extracted from timestamp/datetime, so drop timestamp/datetime.

df = df.drop(columns=['timestamp', 'datetime'])
```

```python
# Divide the variables into numeric and categoric, and analyse accordingly
```

```python
# What is functioning day, and what relationship does it bear to rented bike count?

fig = px.bar(
    df.groupby("Functioning Day")["rented_bike_count"].sum().reset_index(),
    x="Functioning Day",
    y="rented_bike_count",
    title="Total Rented Bike Count by Functioning Day",
    labels={"rented_bike_count": "Total Bike Count", "Functioning Day": "Functioning Day"}
)

fig.show()
```

```python
# Functioning Day determines whether the bike service is running - so needs to be removed.
# Drop all cases where Functioning Day = No. Then delete Functioning Day variable.

print('shape before modification', df.shape)
df = df[df["Functioning Day"] != "No"]
df = df.drop(columns=["Functioning Day"])
print('shape after modification', df.shape)
```

```python
# Save as csv in Drive
df.to_csv('data/interim/02-data-after-EDA.csv', index=False)
```
